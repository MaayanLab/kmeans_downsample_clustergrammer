# K-means Downsample Clustergrammer Example

```python
from clustergrammer_widget import *
net = Network(clustergrammer_widget)
```

# Visualize Original Data
First we will visualize an example dataset.

```python
net.load_file('rc_two_cats.txt')
net.cluster()
net.widget()
```

Make a copy of the rows before performing downsampling

```python
df = net.export_df()
print(df.shape)
rows = df.index.tolist()
print(rows)
```

# Perform K-means Downsampling
Here we will perform K-means downsampling on the rows to form 5 clusters. K-means downsampling returns a Numpy array (referred to as `ds_data` below) the length of the original rows/columns that contains the cluster assignment for each row/column.

```python
ds_data = net.downsample(axis='row', ds_type='kmeans', num_samples=5)
```

```python
print(len(ds_data))
print(ds_data)
```

Below we visualize the downsampled heatmap

```python
net.cluster()
net.widget()
```

# Cluster-2

We see that `cluster-2` has only one row. The only row in `cluster-2` can be found at `ds_data[28]` which corresponds to 'ROS1` (see below):

```python
ds_data[28]
```
